# Backprop algoritam

```python
class mulNode:
    def forward(self, a, b):
        self.a = a
        self.b = b
        c = a*b
        return c
    
    def backward(self, dc):
        da = self.b*dc
        db = self.a*dc
        return [da, db]
    
g = mulNode()
print g.forward(2,3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = np.array([1, 2, 3, 4, 5, 6, 7]).astype(np.float32)
c = np.array([2.1, 3.9, 5.7, 8.01, 10.1, 11.9, 13.7]).astype(np.float32)

fig, (ax1) = plt.subplots(1, 1)
plt.subplots_adjust(wspace=.8)
fig.set_size_inches(10, 4)
ax1.scatter(a, c, alpha=.7,c='r')
plt.show()
```

```python
fig, (ax1) = plt.subplots(1, 1)
plt.subplots_adjust(wspace=.8)
fig.set_size_inches(10, 4)
ax1.scatter(a, c, alpha=.7,c='r')

b=0.1

line_x_range = (0, 8)
ax1.plot(line_x_range, [b*x for x in line_x_range], "g", alpha=0.6)
plt.show()
```

```python
fig, (ax1) = plt.subplots(1, 1)
plt.subplots_adjust(wspace=.8)
fig.set_size_inches(10, 4)
ax1.scatter(a, c, alpha=.7,c='r')

b=0.1
lr = 0.01

for it in range(0, 1000):
    for i in range(0, len(a)):
        A = a[i]
        C = c[i]
        CC = g.forward(A, b)
        dc = CC-C
        [da, db]=g.backward(dc)
        b -= lr*db
print(b)

line_x_range = (0, 8)
ax1.plot(line_x_range, [b*x for x in line_x_range], "g", alpha=0.6)
plt.show()
```

```python
class addNode:
    def forward(self, a, b):
        self.a = a
        self.b = b
        c = a+b
        return c
    
    def backward(self, dc):
        da = 1*dc
        db = 1*dc
        return [da, db]
    
h = addNode()
print h.forward(2,3)
```

```python
b=0.1
d=0.1

for it in range(0, 1000):
    for i in range(0, len(a)):
        A = a[i]
        C = c[i]
        # forward
        G = g.forward(A, b)
        CC = h.forward(G, d)
        # backward
        dc = CC-C        
        [dg, dd]=h.backward(dc)
        [da, db]=g.backward(dg)
        # corrections
        b -= 0.001*db
        d -= 0.001*dd
print(b, d)
```
